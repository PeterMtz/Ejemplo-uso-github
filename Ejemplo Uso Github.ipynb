```python

import pandas as pd

# Crear un DataFrame de ejemplo
df = pd.DataFrame({
    'fecha': pd.date_range(start='2024-01-01', periods=10, freq='D'),
    'ventas': [10, 13, 15, 18, 17, 20, 19, 22, 23, 25]
})

df.head()
```

```python

import matplotlib.pyplot as plt
plt.figure(figsize=(15, 6))
plt.plot(df['fecha'], df['ventas'], marker='o')
plt.title('Ejemplo de Serie de Tiempo: Ventas Diarias')
plt.xlabel('Fecha')
plt.ylabel('Ventas')
plt.grid(True)
plt.show()
```
